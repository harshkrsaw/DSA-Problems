# Implement a Queue

It's very common to be asked to implement a Queue class! The class should be able to do the following:

* Check if Queue is Empty
* Enqueue
* Dequeue
* Return the size of the Queue

```python
class Queue():
    
    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return len(self.items) == 0
        
    def Enqueue(self,item):
        self.items.append(item)
        
    def Dequeue(self):
        return self.items.pop(0)
        
    def size(self):
        return len(self.items)            
```

```python
q = Queue()
```

```python
q.isEmpty()
```

```python
q.Enqueue(3)
```

```python
q.Dequeue()
```

```python
q.size()
```

# Implement a Queue

It's very common to be asked to implement a Queue class! The class should be able to do the following:

* Check if Queue is Empty
* Enqueue
* Dequeue
* Return the size of the Queue

```python
class Queue(object):
    def __init__(self):
        self.item = []
    def isEmpty(self):
        return len(self.item) == 0
    def enqueue(self,item):
        self.item.insert(0,item)
    def dequeue(self):
        return self.item.pop()
    def size(self):
        return len(self.item)
```

```python
q = Queue()
```

```python
q.enqueue(4)
```

```python
q.size()
```

```python
q.dequeue()
```
