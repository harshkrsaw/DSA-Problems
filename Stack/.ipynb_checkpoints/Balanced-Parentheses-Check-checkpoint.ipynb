## Problem Statement

Given a string of opening and closing parentheses, check whether it’s balanced. We have 3 types of parentheses: round brackets: (), square brackets: [], and curly brackets: {}. Assume that the string doesn’t contain any other character than these, no spaces words or numbers. As a reminder, balanced parentheses require every opening parenthesis to be closed in the reverse order opened. For example ‘([])’ is balanced but ‘([)]’ is not. 


You can assume the input string has no spaces.

```python
def balance_check(s):
    
    if len(s)%2 != 0:
        return False
        
    stack = []
    openings = set('{[(')
    matches = set([('{','}'),('[',']'),('(',')')])
    
    for ele in s:
        
        if ele in openings:
            stack.append(ele)
        else:
            if len(stack) == 0:
                return False
            
            last = stack.pop()

            if (last,ele) not in matches:
                return False
                
    return len(stack) == 0
```

```python
balance_check('[]')
```

```python
balance_check('[](){([[[]]])}')
```

```python
balance_check('[](){([[[]]])}(')
```

## Problem Statement

Given a string of opening and closing parentheses, check whether it’s balanced. We have 3 types of parentheses: round brackets: (), square brackets: [], and curly brackets: {}. Assume that the string doesn’t contain any other character than these, no spaces words or numbers. As a reminder, balanced parentheses require every opening parenthesis to be closed in the reverse order opened. For example ‘([])’ is balanced but ‘([)]’ is not. 


You can assume the input string has no spaces.

```python
def parenthesis_check(s):

    if len(s) % 2 != 0:
        return False
    
    opening = set('{[(')
    matches = set([('{','}'),('[',']'),('(',')')])
    stack = []
    
    for ele in s:
        
        if ele in opening:
            stack.append(ele)
            
        else:
            
            if len(stack) == 0:
                return False
                
            temp = stack.pop()
            if (temp,ele) not in matches:
                return False
                
    return len(stack) == 0
          
```

```python
parenthesis_check('[([[]({){]]})}]')
```

```python
parenthesis_check('[](){([[[]]])}(')
```

```python
parenthesis_check('[](){([[[]]])}')
```
