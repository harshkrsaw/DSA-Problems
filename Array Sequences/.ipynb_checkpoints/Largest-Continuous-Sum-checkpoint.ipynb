# Largest Continuous Sum

## Problem
Given an array of integers (positive and negative) find the largest continuous sum.

Input:
    [1,2,-1,3,4,10,10,-10,-1]

Output:
    29


## Solution

Fill out your solution below:

```python
def large_cont_sum(arr):
    if len(arr) == 0:
        return 0
    
    current_sum = max_sum = arr[0]
    for num in arr[1::]:
        current_sum = max(num,current_sum + num)
        max_sum = max(current_sum, max_sum)
    return max_sum
```

```python
large_cont_sum([1,2,-1,3,4,10,10,-10,-1])
```

```python
large_cont_sum([1,2,-1,3,4,-1])
```

```python
large_cont_sum([-1,1])
```

# Largest Continuous Sum

## Problem
Given an array of integers (positive and negative) find the largest continuous sum.

Input:
    [1,2,-1,3,4,10,10,-10,-1]

Output:
    29


## Solution

Fill out your solution below:

```python
def cont_largest_sum(arr):

    if len(arr) == 0:
        return 0
    
    current_sum = total_sum = arr[0]

    for num in arr[1::]:
        current_sum = max(num,num+current_sum)
        total_sum = max(current_sum,total_sum)
    return total_sum
    
```

```python
cont_largest_sum([1,2,-1,3,4,-1])
```

```python
cont_largest_sum([-1,1])
```

```python
cont_largest_sum([1,2,-1,3,4,10,10,-10,-1])
```
