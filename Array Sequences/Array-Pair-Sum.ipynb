# Array Pair Sum

## Problem

Given an integer array, output all the ** *unique* ** pairs that sum up to a specific value **k**.

So the input:
    
    pair_sum([1,3,2,2],4)

would return **2** pairs:

     (1,3)
     (2,2)

**NOTE: FOR TESTING PURPOSES CHANGE YOUR FUNCTION SO IT OUTPUTS THE NUMBER OF PAIRS**

```python
def Array_Pair_Sum(arr,k):
    seen = set()
    output = set()
    
    for num in arr:
        temp = k - num

        if temp in seen:
            output.add((min(num,temp),max(num,temp)))
        else:
            seen.add(num)
            
    return len(output)    
```

```python
#CHECK

Array_Pair_Sum([1,9,2,8,3,7,4,6,5,5,13,14,11,13,-1],10)
```

## Problem

Given an integer array, output all the ** *unique* ** pairs that sum up to a specific value **k**.

So the input:
    
    pair_sum([1,3,2,2],4)

would return **2** pairs:

     (1,3)
     (2,2)

**NOTE: FOR TESTING PURPOSES CHANGE YOUR FUNCTION SO IT OUTPUTS THE NUMBER OF PAIRS**

```python
def sum_pair_check(lst,k):
    temp = set()
    output = set()
    for num in lst:
        x = k - num
        if x not in temp:
            temp.add(num)
        else:
            output.add((min(x,num),max(x,num)))
    return len(output)
    
```

```python
sum_pair_check([1,3,2,2],4)
```
