# Fibonnaci Sequence

## Problem Statement

Implement a [Fibonnaci Sequence](https://en.wikipedia.org/wiki/Fibonacci_number) in three different ways:

* Iteratively
* Recursively
* Dynamically (Using Memoization to store results)


Remember that a fibonacci sequence: 0,1,1,2,3,5,8,13,21,...

```python
def fib_iterative(n):
    
    if n <= 0:
        return n
        
    a = 0
    b = 1

    for num in range(n):
        a, b = b, a+b

    return a
```

```python
fib_iterative(10)
```

```python
def feb_rec(n):

    if n == 0 or n == 1:
        return n

    else:
        return feb_rec(n-1) + feb_rec(n-2)
```

```python
feb_rec(10)
```

```python
def feb_dyn(n, memo={}):

    if n in memo:
        return memo[n]
    if n <= 1:
        return n

    memo[n] = feb_dyn(n-1, memo)+ feb_dyn(n-2, memo)
    return memo[n]
```

```python
feb_dyn(10)
```

# Fibonnaci Sequence

## Problem Statement

Implement a [Fibonnaci Sequence](https://en.wikipedia.org/wiki/Fibonacci_number) in three different ways:

* Iteratively
* Recursively
* Dynamically (Using Memoization to store results)


```python
def fib_iter(n):
    
    if n <= 0:
        return n
        
    a = 0
    b = 1

    for num in range(n):
        a, b = b, a+b

    return a
    
```

```python
fib_iter(7)
```

```python
def fib_rec(n):
    
    if n <= 1:
        return n
    else:
        return fib_rec(n-1) + fib_rec(n-2)
```

```python
def fib_dyn(n, memo = {}):
    
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib_dyn(n-1, memo) + fib_dyn(n-2, memo)
    return memo[n]
    
```

```python
fib_dyn(10)
```

```python
#Another dyn Way(more efficient)

def fib_dyn(n, memo=None):
    if memo is None:
        memo = {}
        
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib_dyn(n-1, memo) + fib_dyn(n-2, memo)
    return memo[n]
```

```python
fib_dyn(10)
```

```python
fib_dyn(0)
```
