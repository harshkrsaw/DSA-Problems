# Fibonnaci Sequence

## Problem Statement

Implement a [Fibonnaci Sequence](https://en.wikipedia.org/wiki/Fibonacci_number) in three different ways:

* Recursively
* Dynamically (Using Memoization to store results)

Remember that a fibonacci sequence: 0,1,1,2,3,5,8,13,21,...

```python
def feb_rec(n):

    if n == 0 or n == 1:
        return n

    else:
        return feb_rec(n-1) + feb_rec(n-2)
```

```python
feb_rec(10)
```

```python
def feb_dyn(n, memo={}):

    if n in memo:
        return memo[n]
    if n <= 1:
        return n

    memo[n] = feb_dyn(n-1, memo)+ feb_dyn(n-2, memo)
    return memo[n]
```

```python
feb_dyn(10)
```
