# Sequential Search


## Unordered list

```python
def Seq_search_unordered(arr,ele):
    output = False
    i = 0
    while i < len(arr) and not output:

        if arr[i] == ele:
            output = True
        else:
            i += 1
    return output
```

## Ordered List

```python
def Seq_search_ordered(arr, ele):
    output = False
    i = 0
    stopped = False
    while i < len(arr) and not output and not stopped:

        if arr[i] == ele:
            output = True
            
        else:
            if arr[i] < ele:
                i += 1
                stopped = True
            else:
                i += 1
                
    return output
```
