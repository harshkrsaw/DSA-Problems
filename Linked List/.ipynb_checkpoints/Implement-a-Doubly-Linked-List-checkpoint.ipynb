# Implement a Doubly Linked List

For this interview problem, implement a node class and show how it can be used to create a doubly linked list.

```python
class Node(object):
    def __init__(self,value):
        self.value = value
        self.nextnode = None
        self.prevnode = None
```

```python
# Create a Doubly Linked List here

a = Node(1)
b = Node(3)
c = Node(5)
d = Node(7)

a.nextnode = b
b.prevnode = a 
b.nextnode = c
c.prevnode = b
c.nextnode = d
d.prevnode = c
```
