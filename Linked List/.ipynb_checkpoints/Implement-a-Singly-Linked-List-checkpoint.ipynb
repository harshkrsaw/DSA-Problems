# Implement a Singly Linked List

For this interview problem, create a node class and show how it can be used to create a Singly Linked List

```python
class Node(object):
    
    def __init__(self,value):
        self.value = value
        self.nextnode = None
```

```python
# Create a Singly Linked List here

a = Node(1)
b = Node(3)
c = Node(5)
d = Node(7)

a.nextnode = b
b.nextnode = c
c.nextnode = d
```

```python
a.nextnode.value
```
